```python
from sklearn.cluster import KMeans
```

```python
newdf=pd.read_csv('/content/Mydrive/My Drive/new_results.csv') #made pandas dataframe for the dataset
```

```python
from sklearn.cluster import KMeans
```

```python
error = []
K = range(1,10)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(newdf)
    error.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, error, 'bx-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Error')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# run kmeans 3 times 
kmeans = KMeans(n_clusters=3)
kmeans.fit(newdf)
kmeans.labels_ # labeling them
predictions = kmeans.predict(newdf) #predicting the cluster for all rows
```

```python
# labeling the rows of dataset and add column 'cluster' in the dataset
newdf['cluster']=kmeans.labels_
newdf
```
